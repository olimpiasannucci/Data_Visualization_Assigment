```python
#libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#imoport the dataset
df = pd.read_csv(r"C:\Users\osann\PycharmProjects\datavis\healthy_lifestyle_city_2021.csv")
df
```

```python
#Rename columns
df.rename(columns = {"Sunshine hours(City)":"sunhrs"}, inplace = True)
df.rename(columns = {"Cost of a bottle of water(City)":"costbottle_w"}, inplace = True)
df.rename(columns = {"Obesity levels(Country)":"obesity"}, inplace=True)
df.rename(columns = {"Life expectancy(years) (Country)":"life_exp"}, inplace = True)
df.rename(columns = {"Pollution(Index score) (City)":"pollution"}, inplace = True)
df.rename(columns = {"Annual avg. hours worked":"annual_hrswork"}, inplace = True)
df.rename(columns = {"Happiness levels(Country)":"happiness"},inplace = True)
df.rename(columns = {"Outdoor activities(City)":"out_act"}, inplace = True)
df.rename(columns = {"Number of take out places(City)": "takeout_plc"}, inplace = True)
df.rename(columns = {"Cost of a monthly gym membership(City)":"cost_gym"}, inplace = True)
df.head()
```

```python
df.info()
```

```python
#converting object values to float
df['costbottle_w'] = df['costbottle_w'].str[1:].astype('float64')
df['obesity'] = df['obesity'].str[:-1].astype('float64')
df['cost_gym'] = df['cost_gym'].str[1:].astype('float64')
df.info()
```

```python
#checking for NaN values and replacing them with the mean
df.replace('-',np.NaN,inplace=True)

missing_value = df.isnull()
missing_value.head()
for column in missing_value.columns.values.tolist():
    print(column)
    print(missing_value[column].value_counts())
    print(" ")
    
df.isna().sum()
```

```python
df['pollution'] = df['pollution'].astype('float64')
mean_pollution = df['pollution'].mean()
df["pollution"] = df["pollution"].replace(np.nan, mean_pollution)
df['sunhrs'] = df['sunhrs'].astype('float64')
df['annual_hrswork'] = df['annual_hrswork'].astype('float64')
df.info()
```

```python
mean_annualhrswork = df['annual_hrswork'].mean()
df["annual_hrswork"] = df["annual_hrswork"].replace(np.nan, mean_annualhrswork)

mean_sunhrs = df['sunhrs'].mean()
df["sunhrs"] = df["sunhrs"].replace(np.nan, mean_sunhrs)
```

'NaN' values have been replaced with their mean, since it is quite difficult to process the data having 'NaN'. 

```python
#Correlation Matrix

plt.figure(figsize=(15, 7))
plt.title('Correlation Matrix', size = 18, fontweight="bold")
corr = df.corr()
cmap = sns.diverging_palette(220, 20, as_cmap=True)
sns.heatmap(corr,
        xticklabels=corr.columns,
        yticklabels=corr.columns, cmap = cmap, annot=True)

```

# Results of the Correlation Matrix
- Happiness, cost of water bottle and life expectancy are variables that are strongly positively correlated
- while happiness and the cost of water of bottle are strongly negatively correlated with the level of pollution; an increase of the latter will have a negative impact on happiness and the cost of water bottles. 
- other variables' correlations are not relevant since their value is close to 0

```python
plt.figure(figsize=(15, 7))
plt.title('Happiness level for each city', size = 16, fontweight="bold")
plt.xticks(rotation = 85)
plt.xlabel("Cities", size = 16)
plt.ylabel("Happiness", size = 16)

sns.barplot(x=df['City'], y=df['happiness'],
            palette=["darkblue" if x == 'Milan' else 'lightgray' for x in df.City])
```

```python
#Number of take out places according to city's first letter - bar chart
df["City's initial letter"] = 'A-D'

#Alphabetical range
def range_char(start, stop):
    return (chr(n) for n in range(ord(start.upper()), ord(stop.upper()) + 1))

#Adding a column of a dataset with the city's initial letter ranges
for city in df['City']:
    if city[0] in range_char("E", "H"):
        i1 = df[df['City'] == city].index[0]
        df.loc[i1, "City's initial letter"] = 'E-H'
    elif city[0] in range_char("I", "L"):
        i2 = df[df['City'] == city].index[0]
        df.loc[i2, "City's initial letter"] = 'I-L'
    elif city[0] in range_char("M", "P"):
        i3 = df[df['City'] == city].index[0]
        df.loc[i3, "City's initial letter"] = 'M-P'
    elif city[0] in range_char("Q", "U"):
        i4 = df[df['City'] == city].index[0]
        df.loc[i4, "City's initial letter"] = 'Q-U'
    elif city[0] in range_char("V", "Z"):
        i5 = df[df['City'] == city].index[0]
        df.loc[i5, "City's initial letter"] = 'V-Z'

df_barchart = df.groupby("City's initial letter").sum().reset_index()
df = df.drop(["City's initial letter"], axis= 1)

sns.set_theme(style="whitegrid")
bc = sns.barplot(x="City's initial letter", y="takeout_plc", data=df_barchart, color = "darkmagenta")
bc.set_title('Number of take out places - Sum by groups of cities',fontweight="bold", size=14)
bc.set_xlabel("City's initial letter", fontweight="bold")
bc.set_ylabel('Number of take out places', fontweight="bold")
plt.show()
```

```python
f2 = plt.figure(figsize=(25, 16))
plt.tight_layout()
gs = f2.add_gridspec(3, 4)

#Chart 1: Life Expectancy and Obesity
f2_ax1 = f2.add_subplot(gs[0, :1])
sns.scatterplot(x = 'life_exp', y='obesity', data = df, s = 40, color = 'mediumblue')
f2_ax1.set_title('Life expectancy (years) vs Obesity (%)', size = 17, fontweight="bold")
plt.xlabel('Life Expectancy', size = 17)
plt.ylabel('Obesity', size = 17)

#Chart 2: Bar Plot of Life expectancy with the bars in a non-increasing length order
f2_ax2 = f2.add_subplot(gs[0,1:-1])
sns.barplot(x ='City', y='life_exp', data = df, 
            order=df.sort_values('life_exp',ascending = False).City, palette = "rocket_r")
plt.xticks(rotation = 70)
plt.tight_layout()
plt.ylabel('Life Expectancy (years)')
f2_ax2.set_title('Life Expectancy for each city', size = 17, fontweight="bold")

#Chart 3: Heatmap that shows the correlations between the last 5 variables
f2_ax3 = f2.add_subplot(gs[0:2, -1])
df2 = df.iloc[:,-5:]
corr = df2.corr()
cmap = sns.diverging_palette(250, 20, as_cmap=True)
sns.heatmap(corr,cmap = cmap, annot=True)
f2_ax3.set_title('Correlations between the last five variables', size = 17, fontweight="bold")

#Chart 4: City with Life Expectancy < 75 years 
f2_ax4 = f2.add_subplot(gs[1, 0])
df3 = df[df['life_exp'].values > 75]
sns.scatterplot(x = 'life_exp', y='obesity', data = df3, s = 50, color = 'indigo')
plt.xlabel('Life Expectancy', size = 17)
plt.ylabel('Obesity', size = 17)
f2_ax4.set_title('Life Expectancy > 75 (years) vs Obesity(%)', size = 17, fontweight="bold")


#Chart 5: Pie Chart representing the pollution level
f2_ax5 = f2.add_subplot(gs[1:, 1])
df['pollution'] = df['pollution'].astype('int')
df['pollution level'] = '0 - 20'
for pollution in df['pollution']:
    if int(pollution) in range(20,40):
        i2 = df[df['pollution'] == pollution].index[0]
        df.loc[i2, 'pollution level'] = '20 - 40'
    elif int(pollution) in range(40,60):
        i3 = df[df['pollution'] == pollution].index[0]
        df.loc[i3, 'pollution level'] = '40 - 60'
    elif int(pollution) in range(60,80):
        i4 = df[df['pollution'] == pollution].index[0]
        df.loc[i4, 'pollution level'] = '60 - 80'
    elif int(pollution) in range(80,100):
        i5 = df[df['pollution'] == pollution].index[0]
        df.loc[i5, 'pollution level'] = '80 - 100'

df.groupby('pollution level').size().plot(kind='pie', autopct='%1.1f%%', textprops={'fontsize': 17}, 
                                          colors = ['navajowhite','khaki','lightseagreen','lightgreen','lightcoral'],).set_ylabel('')
                                                             
f2_ax5.set_title('Pollution level',fontweight="bold", size=18)
df = df.drop(["pollution level"], axis= 1)    
   
 
#Chart 6: Sunshine Hours vs Outdoor Activities
f2_ax6 = f2.add_subplot(gs[1, 2])
sns.scatterplot(x = 'out_act', y='sunhrs', data = df3, s = 40, color = 'orangered')
plt.xlabel('Outdoors Activities', size = 16)
plt.ylabel('Sunshine Hours', size = 16)
f2_ax6.set_title('Sunshine Hours vs Outdoors Activities', size = 15, fontweight="bold")

#Chart 7: Life Expectancy < 80 vs Obesity < 25%
f2_ax7 = f2.add_subplot(gs[2,0])
df4 = df[df['life_exp'].values < 80]
df4 = df4[df4['obesity'].values < 25]
sns.scatterplot(x = 'life_exp', y='obesity', data = df4, s = 40, color = 'teal')
plt.xlabel('Life Expectancy', size = 16)
plt.ylabel('Obesity', size = 16)
f2_ax7.set_title('Life Expectancy < 80 vs Obesity < 25%', size = 16, fontweight="bold")

#Chart 8: Bar Plot of the Gym Costs for each city in decreasing order
f2_ax8 = f2.add_subplot(gs[2, 2:])
sns.barplot(x ='City', y='cost_gym', data = df, 
            order=df.sort_values('cost_gym',ascending = False).City, palette = "viridis")
plt.xticks(rotation = 70)
plt.ylabel('Cost of Gym', size = 18)
f2_ax8.set_title('Cost of Gym for each City', size = 18, fontweight="bold")



```
